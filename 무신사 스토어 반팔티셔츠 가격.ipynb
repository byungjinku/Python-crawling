```python
from bs4 import BeautifulSoup
import csv
import requests
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
import os
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.106 Safari/537.36'}
```

```python
product_name = '반팔티셔츠'
# https://store.musinsa.com/app/items/lists/001001
url = f"https://search.musinsa.com/search/musinsa/?q={product_name}&type="
```

```python
response = requests.get(url, headers = headers)
response
```

```python
#####
## 크롬창을 직접
options = Options()
options.headless = True
browser = webdriver.Chrome(options=options)
browser.implicitly_wait(5)
browser.get(url)
time.sleep(2)
# BeautifulSoup을 활용하여 데이터 파싱
html = browser.page_source
####
```

```python
# BeautifulSoup을 활용하여 데이터 파싱
soup = BeautifulSoup(response.text, "html.parser")
print(soup)
```

```python
#상품명
# <div class="name">반팔티셔츠</div>
names = soup.findAll('p', {'class':'item_title'})

print(len(names))
for item in names :
    item_name = item.text.split()
    print(item.text)
 
```

```python
prices = soup.findAll('p', {'class':'price'})
print(len(prices))
for item in prices :
    item_name = item.text.split()
    print(item_name)
#     if len(item_name) == 1 : 
#         print('정상가')
#     else len(item_name) == 2 :        
#         print('할인가')
# print(len(prices))
```

```python
name_list = []
price_list = []
```

```python
i = 0
for item in names :
    if(i > 0) : 
        name_list.append(item.text)        
    i = i + 1
    
for item in prices :
    price_list.append(item.text)
```

```python
result_file_name = "tshirt.csv"
for k in range(len(price_list)) :
#     print("*상품명 : ", name_list[k+1], ",    *가격 : ",price_list[k])
    result = pd.DataFrame(data={'상품명' : name_list, '가격' : price_list})
    result.to_csv(result_file_name, header = False, index = False, encoding='utf-8-sig')
```
